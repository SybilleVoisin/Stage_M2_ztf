```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import gamma

# Définir la fonction gaussienne
def gaussian(x, amplitude, mean, stddev):
    return amplitude * np.exp(-((x - mean) ** 2) / (2 * stddev ** 2))

# Définir la fonction Moffat
def moffat(x, amplitude, mean, alpha, beta):
    return amplitude * (1 + ((x - mean) ** 2) / (alpha ** 2)) ** (-beta)

# Générer les données
x = np.linspace(-5, 5, 400)


# Paramètres pour la fonction gaussienne
amplitude_gauss = 1
mean_gauss = 0
stddev_gauss = 2

# Paramètres pour la fonction Moffat
amplitude_moffat = 1
mean_moffat = 0
alpha_moffat = 3
beta_moffat = 2

# Calculer les valeurs des fonctions
y_gauss = gaussian(x, amplitude_gauss, mean_gauss, stddev_gauss)
y_moffat = moffat(x, amplitude_moffat, mean_moffat, alpha_moffat, beta_moffat)

# Tracer les fonctions
fig, ax = plt.subplots()
ax.plot(x, y_gauss, label='Gaussian', color='blue', linewidth = 0.5)
ax.plot(x, y_moffat, label='Moffat', color='red', linewidth = 0.5)
ax.set_xlabel('x')
ax.set_yscale("log")
ax.set_ylabel('y')
ax.legend()
```
