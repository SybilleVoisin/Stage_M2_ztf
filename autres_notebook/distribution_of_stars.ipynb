```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#fichier basé sur des catalogues intermédiaires d’ubercal.
#En plus de l’isolation, il y a déjà une sélection importante basée sur gaia 
#(pour ne garder que les étoiles de la main sequence et les WD, avec des cuts maison) plus un cut à PS1_g<19.
df1= pd.read_parquet('count_gaia_filters.parquet', engine='fastparquet')
df1
```

```python
#application des cuts pour retirer les étoiles qui ont plus de 3% de nan ou 1% de flux neg.
mean_stars_per_quadrant = df1.groupby('rcid')['n_star'].mean()
plt.figure(figsize=(10, 6))
plt.bar(mean_stars_per_quadrant.index, mean_stars_per_quadrant.values, color='skyblue')
plt.xlabel('Quadrant number')
plt.ylabel('Average number of stars')
plt.title('Average star distribution by quadrant for the "count_gaia_filters" data')
plt.grid(True)
plt.show()
```

```python
df1_rcid_30 = df1[df1['rcid'] == 30]
plt.figure(figsize=(10, 6))
plt.hist(df1_rcid_30['n_star'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Number of stars')
plt.ylabel('Frequency')
plt.title('Histogram of number of stars for rcid = 30 for the "count_gaia_filters" data')
plt.grid(True)
plt.show()
```

```python
df1_rcid_59 = df1[df1['rcid'] == 59]
plt.figure(figsize=(10, 6))
plt.hist(df1_rcid_59['n_star'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Number of stars')
plt.ylabel('Frequency')
plt.title('Histogram of number of stars for rcid = 59 for the "count_gaia_filters" data')
plt.grid(True)
plt.show()
```
